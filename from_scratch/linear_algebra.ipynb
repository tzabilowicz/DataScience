```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")

# Local Imports
from Mathematics.LinearAlgebra.vectors import (
    Vector,
    add,
    sub,
    mult,
    div,
    dot,
    sum_of_squares,
    scalar_mult,
    scalar_div
)
```

#### Visual Guide to Linear Algebra
This notebook provides a visual guide to linear algebra. Please note, there are no accompanying formulas. It is assumed that you are comfortable with these operations.

```python
"""
First, we will visualize vector operations.
We will start with two arbitrary vectors v1 and v2, and a scalar c.
"""

c: int = 5
v1: Vector = [2, 4]
v2: Vector = [8, 2]
```

```python
# Vector Addition
t = add(v1, v2)

fig, ax = plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="v1")
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label="v2")
ax.quiver(0, 0, t[0], t[1], angles='xy', scale_units='xy', scale=1, color='b', label="sum")

ax.set_xlim([-5, 20])
ax.set_ylim([-5, 20])
plt.title("v1 + v2")
ax.set_aspect("equal")
plt.legend()
plt.show()
```

```python
# Vector Subtraction
t = sub(v1, v2)

fig, ax = plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="v1")
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label="v2")
ax.quiver(0, 0, t[0], t[1], angles='xy', scale_units='xy', scale=1, color='b', label="difference")

ax.set_xlim([-10, 10])
ax.set_ylim([-5, 10])
ax.set_aspect("equal")
plt.title("v1 - v2")
plt.legend()
plt.show()
```

```python
# Vector Multiplication
t = mult(v1, v2)

fig, ax = plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="v1")
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label="v2")
ax.quiver(0, 0, t[0], t[1], angles='xy', scale_units='xy', scale=1, color='b', label="product")

ax.set_xlim([-5, 18])
ax.set_ylim([-5, 15])
ax.set_aspect("equal")
plt.title("v1 * v2")
plt.legend()
plt.show()
```

```python
# Vector Division
t = div(v1, v2)

fig, ax = plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="v1")
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label="v2")
ax.quiver(0, 0, t[0], t[1], angles='xy', scale_units='xy', scale=1, color='b', label="quotient")

ax.set_xlim([-5, 18])
ax.set_ylim([-5, 15])
ax.set_aspect("equal")
plt.title("v1 / v2")
plt.legend()
plt.show()
```

```python
# Scalar Multiplication
t = scalar_mult(c, v1)

fig, ax = plt.subplots()
ax.quiver(0, 0, t[0], t[1], angles='xy', scale_units='xy', scale=1, color='g', label="c*v1")
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="v1")

ax.set_xlim([-5, 18])
ax.set_ylim([-5, 25])
ax.set_aspect("equal")
plt.title(f"{c} * v1")
plt.legend()
plt.show()
```

```python
# Dot Product
d_p = dot(v1, v2)

# Projection
proj_v1_v2 = scalar_mult((d_p/ sum_of_squares(v2)), v2)

fig, ax = plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="v1")
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label="v2")
ax.quiver(0, 0, proj_v1_v2[0], proj_v1_v2[1], angles='xy', scale_units='xy', scale=1, color='b', label="v1 onto v2")

# Show the distance
ax.plot([v1[0], proj_v1_v2[0]], [v1[1], proj_v1_v2[1]], 'k--', alpha=0.3)

ax.set_xlim([-5, 10])
ax.set_ylim([-5, 10])
ax.set_aspect("equal")
plt.title(f"Dot Product v1, v2: {d_p}")
plt.legend()
plt.show()
```
