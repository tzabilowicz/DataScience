```python
import pandas as pd
import matplotlib.pyplot as plt

# Settings
plt.style.use('ggplot')
```

```python
# Read data
hot_wings = pd.read_csv("../Data/Beerwings.csv")

# Male metadata
males = hot_wings[hot_wings["Gender"] == "M"]
print("Males:")
print(males.describe())

# Female metadata
females = hot_wings[hot_wings["Gender"] == "F"]
print("Females:")
print(females.describe())
```
