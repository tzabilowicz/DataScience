```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
# Distribution parameters
mu = 23
var = 49

B = 1000
S = 50
```

```python
# Single sample
sample = np.random.normal(mu, var, S)

print(f"Sample Mean: {np.mean(sample)}")
print(f"Sample Std:  {np.std(sample)}")
```

```python
bootstrap_mu = []
for b in range(B):
    sample = np.random.normal(mu, var, S)
    bootstrap_mu.append(np.mean(sample))

plt.hist(bootstrap_mu, bins=15, edgecolor="black", alpha=0.75)
plt.show()
```

```python
print(f"Bootstrap Mean: {np.mean(bootstrap_mu)}")
print(f"Bootstrap Std:  {np.std(bootstrap_mu)}")
```
